### Задание 2

С помощью функции `linspace` из библиотеки `Numpy` создайте массив **t** из **51** числа от **0** до **10** включительно.

Создайте массив `Numpy` под названием **f**, содержащий косинусы элементов массива **t**.

Постройте линейную диаграмму, используя массив **t** для координат по горизонтали,а массив **f** - для координат по вертикали.  Линия графика должна быть *зеленого* цвета.

Выведите название диаграммы - *График f(t)*. Также добавьте названия для горизонтальной оси - *Значения t* и для вертикальной - *Значения f*.

Ограничьте график по оси **x** значениями **0.5** и **9.5**, а по оси **y** - значениями **-2.5** и **2.5**.

```python
t = np.linspace(0, 10, 51)  # масив из 51 чисела от 0 до 10 включительно 
f = np.cos(t)  # масив из 51 чисела cos(t)
```

```python
# Задание свойств шрифта для заголовка графика
title_font = {
    "fontsize": 14,
    "fontweight": "bold",
    "color": "#808080",
    "family": "sans",
}

# Задание свойств шрифта для нименования осей графика
label_font = {
    "fontsize": 10,
    "family": "sans",
}
```

```python
plt.plot(t, f, color="green")  # рисование графика

plt.title("График f(t)", fontdict=title_font)  # наименование графика
plt.xlabel("Значения t", fontdict=label_font)  # наименование оси х
plt.ylabel("Значения f", fontdict=label_font)  # наименование оси y

plt.axis([0.5, 9.5, -2.5, 2.5])  # задание границ отображения по оси х и оси y 

plt.grid(axis="y", color="ghostwhite")  # вывод сетки графика

plt.show()  # отобразить получившийся график
```

### Задание 3*

С помощью функции `linspace` библиотеки `Numpy` создайте массив **x** из **51** числа от **-3** до **3** включительно.  Создайте массивы **y1**, **y2**, **y3**, **y4** по следующим формулам:  
$y_1 = x^2$  
$y_2 = 2 * x + 0.5$  
$y_3 = -3 * x - 1.5$  
$y_4 = sin(x)$

Используя функцию `subplots` модуля `matplotlib.pyplot`, создайте объект `matplotlib.figure.Figure` с названием **fig** и массив объектов *Axes* под названием **ax**,причем так, чтобы у вас было **4** отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив **x** используется для координат по горизонтали. В левом верхнем графике для координат по вертикали используйте **y1**,в правом верхнем - **y2**, в левом нижнем - **y3**, в правом нижнем - **y4**.

Дайте название графикам: **График y1**, **График y2** и т.д.

Для графика в левом верхнем углу установите границы по оси **x** от **-5** до **5**.

Установите размеры фигуры **8** дюймов по горизонтали и **6** дюймов по вертикали.

Вертикальные и горизонтальные зазоры между графиками должны составлять **0.3**.

```python
x = np.linspace(-3, 3, 51)
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

print(f'Массив x:\n {x}\n')
print(f'Массив y1:\n {y1}\n')
print(f'Массив y2:\n {y2}\n')
print(f'Массив y3:\n {y3}\n')
print(f'Массив y4:\n {y4}\n')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.4, hspace=0.5)

ax1.plot(x, y1)
ax1.set_title("График y1")
ax1.set_xlabel("Значения x")
ax1.set_ylabel("Значения y1")
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title("График y2")
ax2.set_xlabel("Значения x")
ax2.set_ylabel("Значения y2")

ax3.plot(x, y3)
ax3.set_title("График y3")
ax3.set_xlabel("Значения x")
ax3.set_ylabel("Значения y3")

ax4.plot(x, y4)
ax4.set_title("График y4")
ax4.set_xlabel("Значения x")
ax4.set_ylabel("Значения y4")
```
